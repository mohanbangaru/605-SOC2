```python
X=[]
Z=[]
IMG_SIZE=150

FLOWER_DAISEY_DIR = "C:\\Users\\DELL\\Desktop\\daisy"
FLOWER_SUNFLOWER_DIR ="C:\\Users\\DELL\\Desktop\\sunflower"
FLOWER_TULIP_DIR = "C:\\Users\\DELL\\Desktop\\tulip"
FLOWER_DANDI_DIR ="C:\\Users\\DELL\\Desktop\\dandelion"
FLOWER_ROSE_DIR = "C:\\Users\\DELL\\Desktop\\rose"
```

```python
def assign_label(img, flower_type):
  return flower_type
```

```python
# tqdm --> It creates a progress bar from the loop
!pip install opencv-python
import cv2
# import os
# import numpy as np

# Rest of your code...

def make_train_data(flower_type, DIR):
    for img in tqdm(os.listdir(DIR)):
        label = assign_label(img, flower_type)
        path = os.path.join(DIR, img)
        img = cv2.imread(path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    X.append(np.array(img))
    Z.append(str(label))
```

```python
make_train_data('Daisy', FLOWER_DAISEY_DIR)
print(len(X))
```

```python
make_train_data('Sunflower', FLOWER_SUNFLOWER_DIR)
print(len(X))
```

```python
make_train_data('Tulip', FLOWER_TULIP_DIR)
print(len(X))
```
