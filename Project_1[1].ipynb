```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=2)
```

```python
features=df.iloc[:,2:]
features
```

```python
features.columns=features.columns.astype(str)
features.columns
```

```python
from sklearn.preprocessing import StandardScaler
# StandardScaler ---> is a preprocessing class that is used to standardize or normalize the features of dataset.it scales each features of dataset.it scales each feature in such a way it has a mean 0 and std of 1
scaler=StandardScaler()
scaled_features=scaler.fit_transform(features)
```

```python
scaled_features
```

```python
scaled_features
```

```python
scaled_features.shape
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans
```

```python
df['Cluster']=kmeans.fit_predict(scaled_features)
```

```python
df['Cluster']
```

```python
df
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
kmeans.labels_
```

```python
print(confusion_matrix(df['Cluster'],kmeans.labels_))
```

```python
print(accuracy_score(kmeans.labels_,df['Cluster']))
```
